In "genres" data we only want keywords like "Action", "Adventure", "Fantasy", "Science Fiction"

```python
# Defining a helper function
def convert(obj):
    l = []
    for i in obj:
        l.append(i["name"])
    return l
        
```

convert(movies["genres"])

#In this command, convert is applied directly to the entire "genres" column of the movies DataFrame. 
#This means that the convert function must be designed to handle the entire column as input. 

The data is in the form of a string of lists which we will have to convert to list

```python
np.dtype(movies["genres"])
```

```python
# ast. literal_eval method can safely evaluate strings containing Python values from unknown sources without us having to parse the values.
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

ast.literal_eval successfully converted the input string into a Python list of dictionaries.

```python
def convert(obj):                      # here 'obj' is passed as string
    l = []
    for i in ast.literal_eval(obj):    # here after applying ast.literal_eval() 'obj' is passed as a list
        l.append(i["name"])
    return l
```

```python
movies["genres"].apply(convert)        

# you're applying the apply method to the "genres" column of the movies DataFrame. 
# The apply method iterates over each element (cell) in the "genres" column and applies the convert function to each individual element. 
# This is similar to using a loop to process each element one by one. 
```

```python
# Storing this in the genres column itself
movies["genres"] = movies["genres"].apply(convert) 
```

```python
movies.head()
```

```python
# Same preprocessing to be done for 'keywords' column
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
# For 'cast' column we need only first 3 actor names from the first 3 dictionaries
movies["cast"][0]
```

```python
def convert_cast(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
            counter = counter+1
    return l
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
movies.head()
```

```python
# for 'crew' we want only those values where job is "Director"
movies["crew"][0]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

To find the similarities between movies we will have to convert the text into vectors.

Text vectorization technique "Bag of words"
In "Bag of words" Technique we combine all the tags from all the movies to create a large text.
From this large text we want 5000 most common words

suppose we have 5000 movies ans each movie has text of 100 words then the total Bag of Words will be 500000
Now for each movie we will create a vector based on how many times the most common 500 words appear in each movie.
So each movie is converted to a vector in a 5000-dimensional space.

During the process of Vectorization we will need to remove the stop words like "are, the, in",etc that dont contribute to the meaning of a sentence.
This can be done by using the class CountVectorizer. Countvectorizer is a method to convert text to numerical data.


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df["tags"])
```

```python
vectors
```

This implementation produces a sparse representation of the counts, hence, we need to convert these into numpy array

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

there are many similar meaning words like ["loving","loved","love"] and we need to use stemming to convert them to ["love","love","love"]. 
        Stemming is the process of producing morphological variants of a root/base word. 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))
    return " ".join(x)
```

```python
ps.stem("dance","dancing")
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
# So instead we use enumerate(). 
# enumerate function is a built-in function that allows you to iterate through a sequence and keep track of the index of each element. 
list(enumerate(similarity[3]))
```

```python
sorted(list(enumerate(similarity[3])),reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print (i)
```
